```python
#items which have protein above 15, sugar below 15 and cholesterol below 70
filtered_data = data[(data['Sugars'] < 15) & (data['Cholesterol'] < 70) & (data['Protein'] > 15) & (data['Calories'] <= 500)]

# Sort the filtered data by protein levels in descending order
top_10_items = filtered_data.sort_values(by='Protein', ascending=False).head(10)

print("Top 10 Items with Low Sugar and Cholesterol, High Protein, and Maximum Calories of 500:")
print(top_10_items[['Category', 'Item', 'Sugars', 'Cholesterol', 'Protein', 'Calories']])
```
